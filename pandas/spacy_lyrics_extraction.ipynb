```python
tokens = []
tags = []
pos = []

for doc in nlp.pipe(lyrics_df['Text'].astype('unicode').values, batch_size=4341,
                        n_threads=3):
    if doc.is_parsed:
        
        tokens.append([n.text for n in doc])
        tags.append([n.tag_ for n in doc])
        pos.append([n.pos_ for n in doc])  

# df_sample['tokens'] = tokens
# df_sample['tags'] = tags
# df_sample['pos'] = pos
```

```python
len(pos)
```

```python
list_of_nouns = []
list_of_verbs = []
list_of_adjectives = []

i = 0
for nlist in pos:
    
    nouns = []
    verbs = []
    adjectives = []
    
    j = 0 
    for n in nlist:
        
        if n == 'NOUN':
            nouns.append(tokens[i][j])

        if n == 'ADJ':
            adjectives.append(tokens[i][j])

        if n == 'VERB':
            verbs.append(tokens[i][j])  

        j += 1
    
    i +=1 
    
    list_of_nouns.append(nouns)
    list_of_adjectives.append(adjectives)
    list_of_verbs.append(verbs)
    
        
        
```

```python
len(list_of_adjectives)
```

```python
artist = df['Artist Name']
track_name = df['Track Name']
text = df['Text'] 
```

```python
tagged_lyrics_df = pd.DataFrame({"Artist Name":artist,
                            "Track Name":track_name, "Text": text,
                            "Tokens": tokens, "Tags": tags, "POS":pos, 
                            "Nouns": list_of_nouns,"Verbs": list_of_verbs, "Adjectives": list_of_adjectives}, columns=['Artist Name', 'Track Name', 'Text', 'Tokens', 'Tags', 'POS', 'Nouns', 'Verbs', 'Adjectives'])
tagged_lyrics_df
```

```python
grouped_tagged_lyrics = tagged_lyrics_df.groupby(['Artist Name', 'Track Name'])
grouped_tagged_lyrics.count().head()
```
