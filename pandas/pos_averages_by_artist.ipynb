```python
import pandas as pd
```

```python
df = pd.read_csv("pos_ratios.csv")
df.head()
```

```python
pos_df = df[["Artist Name", "pos_count", "nouns_count", "verbs_count", "adjs_counts"]]
```

```python
pos_df_grouped = pos_df.groupby(['Artist Name'])
pos_df_grouped.count().head()
```

```python
pos_avg_df = pd.DataFrame(pos_df_grouped["pos_count", "nouns_count", "verbs_count", "adjs_counts"].mean())
pos_avg_df.head()
```

```python
avg_nouns_ratio = pos_avg_df['nouns_count']/pos_avg_df['pos_count'].round()
avg_verbs_ratio = pos_avg_df['verbs_count']/pos_avg_df['pos_count'].round()
avg_adjs_ratio = pos_avg_df['adjs_counts']/pos_avg_df['pos_count'].round()
```

```python
pos_avg_df['avg_noun_ratio'] = avg_nouns_ratio
pos_avg_df['avg_verb_ratio'] = avg_verbs_ratio
pos_avg_df['avg_adj_ratio'] = avg_adjs_ratio
```

```python
pos_avg_df.head()
```

```python
avg_artist_pos_df = pos_avg_df.to_csv("avg_artist_pos.csv", encoding="utf-8", index=False)
```
