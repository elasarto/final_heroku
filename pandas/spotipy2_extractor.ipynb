```python
import pandas as pd
import requests
import json

import spotipy
import spotipy.util as util
from config import CLIENT_ID, CLIENT_SECRET
import random
import pprint
```

```python
token = util.oauth2.SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)

cache_token = token.get_access_token()
spotify = spotipy.Spotify(cache_token)
```

```python
spotify_df = pd.read_csv("spotify_trackids_20.csv")
spotify_df.head()
```

```python
spotify_df.info()
```

```python
track_id = spotify_df['Track ID']
track_id.head()
```

```python
#this loop works

acousticness = []
danceability = []
duration = []
energy = []
track_ids = []
instrumentalness = []
song_key = []
liveness = []
loudness = []
mode = []
speechiness = []
tempo = []
time_sig = []
feature_type = []
valence = []

i = 0

for track in track_id:

    i = 0
    
    features = spotify.audio_features(track)
    
    for feature in features:
    
        try:
            acousticness.append(features[i]['acousticness'])
            danceability.append(features[i]['danceability'])
            duration.append(features[i]['duration_ms'])
            energy.append(features[i]['energy'])
            track_ids.append(features[i]['id'])
            song_key.append(features[i]['key'])
            liveness.append(features[i]['liveness'])
            loudness.append(features[i]['loudness'])
            mode.append(features[i]['mode'])
            speechiness.append(features[i]['speechiness'])
            tempo.append(features[i]['tempo'])
            time_sig.append(features[i]['time_signature'])
            feature_type.append(features[i]['type'])
            valence.append(features[i]['valence'])
            instrumentalness.append(features[i]['instrumentalness'])

            i += 1     

        except(TypeError):
            print("Error Retrieving results for Track" + str(track))
            continue
```

```python
features_df = pd.DataFrame({"acousticness":acousticness,
                            "danceability":danceability, "duration": duration,
                            "energy": energy, "Track ID": track_ids, "song key":song_key, 
                            "liveness": liveness,"loudness": loudness, "mode": mode, 
                            "speechiness": speechiness, "tempo": tempo, "time signature": time_sig, 
                            "type": feature_type, "valence": valence, "instrumentalness": instrumentalness})
features_df.head()
```

```python
features_df.info()
```

```python
combined_df = pd.merge(spotify_df, features_df, on=['Track ID'], how='outer')
combined_df
```

```python
combined_df = combined_df.sort_values('Artist Name', ascending=True )
combined_df
```

```python
combined_df = combined_df.reset_index(drop=True)
combined_df.head()
```

```python
combined_df = combined_df.to_csv("spotify_songs_20.csv", encoding="utf-8", index=False)
```
