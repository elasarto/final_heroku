```python
import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
lyrics_df = pd.read_csv("songs_and_lyrics.csv")
lyrics_df
```

```python
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

for lyric in lyrics_df["Text"]:
    
    sentiment_analysis = analyzer.polarity_scores(str(lyric))

    #print(sentiment_analysis)
    compound_list.append(sentiment_analysis["compound"])
    positive_list.append(sentiment_analysis["pos"])
    negative_list.append(sentiment_analysis["neg"])
    neutral_list.append(sentiment_analysis["neu"])

    #add the sentiment analysis columns into the dataframe
lyrics_df["Compound Score"] = compound_list
lyrics_df["Positive Score"] = positive_list
lyrics_df["Negative Score"] = negative_list
lyrics_df["Neutral Score"] = neutral_list
```

```python
lyrics_df
```

```python
lyrics_grouped = lyrics_df.groupby(['Artist Name'])
lyrics_grouped.count()
```

```python
lyrics_avg_df = pd.DataFrame(lyrics_grouped["acousticness", "danceability", "duration", "energy", "instrumentalness", "liveness",
                                           "loudness", "mode", "song key", "speechiness", "tempo", "time signature", "valence", "Compound Score",
                                           "Positive Score", "Negative Score", "Neutral Score"].mean())
lyrics_avg_df
```

```python
band_analysis_df = lyrics_avg_df.to_csv("band_analysis.csv", encoding="utf-8")
```
