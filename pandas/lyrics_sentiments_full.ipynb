```python
import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
df = pd.read_csv("songdata2.csv")
df.head()
```

```python
lyrics_full_df = df[['artist', 'song', 'text']]
```

```python
lyrics_full_df.head()
```

```python
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

for lyric in lyrics_full_df["text"]:
    
    sentiment_analysis = analyzer.polarity_scores(str(lyric))

    #print(sentiment_analysis)
    compound_list.append(sentiment_analysis["compound"])
    positive_list.append(sentiment_analysis["pos"])
    negative_list.append(sentiment_analysis["neg"])
    neutral_list.append(sentiment_analysis["neu"])

    #add the sentiment analysis columns into the dataframe
lyrics_full_df["compound"] = compound_list
lyrics_full_df["positive"] = positive_list
lyrics_full_df["negative"] = negative_list
lyrics_full_df["neutral"] = neutral_list
```

```python
lyrics_full_df.head()
```

```python
lyrics_analysis_df = lyrics_full_df.to_csv("lyrics_full_tracks_sentiments.csv", encoding="utf-8", index=False)
```

```python
lyrics_full_df.head()
```

```python
lyrics_full_grouped = lyrics_full_df.groupby(['artist', 'song'])
lyrics_full_grouped.count().head()
```

```python
lyrics_full_avg_df = pd.DataFrame(lyrics_full_grouped["text","compound",
                                           "positive", "negative", "neutral"].mean())
lyrics_full_avg_df.head()
```

```python
lyrics_analysis_df2 = lyrics_full_avg_df.to_csv("lyrics_tracks_avg_sentiments.csv", encoding="utf-8")
```

```python
lyrics_full_grouped = lyrics_full_df.groupby(['artist'])
lyrics_full_grouped.count().head()
```

```python
artist_full_avg_df = pd.DataFrame(lyrics_full_grouped["text","compound",
                                           "positive", "negative", "neutral"].mean())
artist_full_avg_df.head()
```

```python
artist_analysis_df2 = artist_full_avg_df.to_csv("artist_full_sentiments.csv", encoding="utf-8")
```
