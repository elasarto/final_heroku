```python
import pandas as pd
import spacy
```

```python
nlp = spacy.load('en')
```

```python
df = pd.read_csv("songdata2.csv")
df.head()
```

```python
df.info()
```

```python
df2 = df[['artist', 'song', 'text']]
df2.head()
```

```python
df_sample = df2.iloc[:5000]
df_sample.head()
```

```python
tokens = []
tags = []
pos = []

for doc in nlp.pipe(df_sample['text'].astype('unicode').values, batch_size=5000,
                        n_threads=3):
    if doc.is_parsed:
        tokens.append([n.text for n in doc])
        tags.append([n.tag_ for n in doc])
        pos.append([n.pos_ for n in doc])
    else:
        tokens.append(None)
        tags.append(None)
        pos.append(None)

# df_sample['tokens'] = tokens
# df_sample['tags'] = tags
# df_sample['pos'] = pos
```

```python
artist = df_sample['artist']
song = df_sample['song']
```

```python
tagged_lyrics = pd.DataFrame({'artist': artist,'song': song,'words': tokens, 'tags': tags, 'pos': pos}, columns=["artist","song","words","tags","pos"])
tagged_lyrics
```
