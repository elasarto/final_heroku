```python
artists = artist_df['unique artist'].tolist()
```

```python
len(artists)
```

```python
track_ids = []
track_names = []
artist_names = []

# names = ["ABBA", "Ace of Base", "Adam Sandler", "Adele", "Aerosmith"]

i = 0

for artist in artists:
    
    i = 0
    
    results = spotify.search(q=artist, limit=20)
    other = results['tracks']['items']
    
    for result in other:
        
        track_ids.append(results['tracks']['items'][i]['id'])
#         print(artist, results['tracks']['items'][i]['name'])
        track_names.append(results['tracks']['items'][i]['name'])
        artist_names.append(artist)

        i += 1
    
    
```

```python
results_df = pd.DataFrame({"Artist Name":artist_names, "Track Name":track_names, "Track ID":track_ids}, columns=["Artist Name", "Track Name", "Track ID"])
results_df.head()
```

```python
results_df = results_df.to_csv("spotify_trackids_20.csv", encoding="utf-8", index=False)
```
