```python
df = df.sort_values('artist', ascending=True )
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
song_data.head()
```

```python
df2 = df[['artist', 'song','text']]
df2.columns = ['Artist Name', 'Track Name', 'Text']
df2.head()
```

```python
test_combined_df = pd.merge(df2, song_data, how='right', on=['Artist Name', 'Track Name'])
test_combined_df
```

```python
test_combined_df.info()
```

```python
final_df = test_combined_df.dropna(how='any') 
final_df
```
