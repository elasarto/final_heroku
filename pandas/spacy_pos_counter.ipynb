```python
import pandas as pd
import spacy

nlp = spacy.load('en')
```

```python
tokens = []
tags = []
pos = []

for doc in nlp.pipe(lyrics_df['Text'].astype('unicode').values, batch_size=4341,
                        n_threads=3):
    if doc.is_parsed:
        
        tokens.append([n.text for n in doc])
        tags.append([n.tag_ for n in doc])
        pos.append([n.pos_ for n in doc])  
```

```python
len(pos)
```

```python
list_of_nouns = []
list_of_verbs = []
list_of_adjectives = []
list_of_nouns_counts = []
list_of_verbs_counts = []
list_of_adjs_counts = []
list_of_pos_counts = []

noun_count = 0
verb_count = 0
adj_count = 0
pos_count = 0

i = 0
for nlist in pos:
    
    nouns = []
    verbs = []
    adjectives = []
    nouns_counts = []
    verbs_counts = []
    adjs_counts = []
    
    list_of_pos_counts.append(len(nlist))
    
    j = 0 
    for n in nlist:
        
        if n == 'NOUN':
            nouns.append(tokens[i][j])
            nouns_counts.append(noun_count)
            noun_count += 1

        if n == 'ADJ':
            adjectives.append(tokens[i][j])
            adjs_counts.append(adj_count)
            adj_count += 1

        if n == 'VERB':
            verbs.append(tokens[i][j])
            verbs_counts.append(verb_count)
            verb_count += 1

        j += 1
    
    i +=1 
    
    list_of_nouns.append(nouns)
    list_of_adjectives.append(adjectives)
    list_of_verbs.append(verbs)
    list_of_nouns_counts.append(len(nouns_counts))
    list_of_verbs_counts.append(len(verbs_counts))
    list_of_adjs_counts.append(len(adjs_counts))
```

```python
artist = lyrics_df['Artist Name']
track_name = lyrics_df['Track Name']
text = lyrics_df['Text'] 
```

```python
pos_df = pd.DataFrame({"Artist Name":artist,
                            "Track Name":track_name, "Text": text,
                            "Tokens": tokens, "POS":pos, 
                            "Nouns": list_of_nouns,"Verbs": list_of_verbs, "Adjectives": list_of_adjectives, "pos_count":list_of_pos_counts,
                      "nouns_count": list_of_nouns_counts, "verbs_count":list_of_verbs_counts, "adjs_counts":list_of_adjs_counts}, columns=['Artist Name', 'Track Name', 'Text', 'Tokens', 'POS', 'Nouns', 'Verbs', 'Adjectives', 'pos_count','nouns_count', 'verbs_count',
                                                                                                                                           'adjs_counts'])
pos_df
```

```python
nouns_ratio = pos_df['nouns_count']/pos_df['pos_count'].round()
verbs_ratio = pos_df['verbs_count']/pos_df['pos_count'].round()
adjs_ratio = pos_df['adjs_counts']/pos_df['pos_count'].round()
```

```python
pos_df['noun_ratio'] = nouns_ratio
pos_df['verb_ratio'] = verbs_ratio
pos_df['adj_ratio'] = adjs_ratio
```

```python
pos_df
```

```python
words = pos_df['Tokens']

list_of_popular_words = []

for word_list in words:

    word_counter = {}
    for word in word_list:
        
        if word in word_counter:
            word_counter[word] += 1
        else:
            word_counter[word] = 1

    popular_words = sorted(word_counter, key = word_counter.get, reverse = True)

    list_of_popular_words.append(popular_words[:20])
#     print(popular_words[:10])

#     len(popular_words)
```
